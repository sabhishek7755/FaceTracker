### Run Augmentation pipeline

```python
#Augmentation using Albumentations
import albumentations as alb
```

```python
augmentor = alb.Compose([alb.RandomCrop(width = 450, height =450), 
                         alb.HorizontalFlip(p=0.5),
                         alb.RandomBrightnessContrast(p=0.2),
                         alb.RandomGamma(p=0.2),
                         alb.RGBShift(p=0.2),
                         alb.VerticalFlip(p=0.5)],bbox_params=alb.BboxParams(format='albumentations', # to fix the annotations
                                                                             label_fields=['class_labels']))
```

```python
for Set in ['train','test','val']:
    for image in os.listdir(os.path.join('data',Set,'images')):
        img = cv2.imread(os.path.join('data', Set,'images',image))
        
        coords = [0,0,0.00001,0.00001] # this is default annotation
        label_path = os.path.join('data',Set,'labels', f'{image.split(".")[0]}.json')
        if os.path.exists(label_path): # check if annotation exist or else default annotaion will be assigned
            with open(label_path,'r') as f:
                label = json.load(f)
            
            coords[0] = label['shapes'][0]['points'][0][0]
            coords[1] = label['shapes'][0]['points'][0][1]
            coords[2] = label['shapes'][0]['points'][1][0]
            coords[3] = label['shapes'][0]['points'][1][1]
            coords = list(np.divide(coords,[640,480,640,480]))
            
        try:
            for x in range(60):
                augmented = augmentor(image = img, bboxes = [coords], class_labels = ['face']) # get augmented image in 450x450 size
                cv2.imwrite(os.path.join('aug_data', Set, 'images', f'{image.split(".")[0]}.{x}.jpg'), augmented['image']) # save tge aug image
                
                #Now save the labels and annotations in new jason file for all images
                annotation = {}
                annotation['image'] = image
                
                if os.path.exists(label_path):
                    if len(augmented['bboxes']) == 0: # jason label file exist but without any bounding box i.e. no face
                        annotation['bbox'] = [0,0,0,0]
                        annotation['class'] = 0 
                    else:
                        annotation['bbox'] = augmented['bboxes'][0] # bounding box of the face
                        annotation['class'] = 1 # has a face
                
                else:
                    annotation['bbox'] = [0,0,0,0]
                    annotation['class'] = 0 # not having face
                
                with open(os.path.join('aug_data',Set, 'labels', f'{image.split(".")[0]}.{x}.json'), 'w') as f:
                    json.dump(annotation, f) # writing the new json file for labels
            
        
        except Exception as e:
            print(e)
        
```

### Prepare Labels

```python
# function to load label from json files in labels folder
def load_labels(label_path): # label path is a tensor object
    with open(label_path.numpy(), 'r', encoding = 'utf-8') as f: # .numpy() is used to convert the tensor object into a string, 'ytf-8 is standard encoding used to represt text
        label = json.load(f)
    return [label['class']], label['bbox']
```

```python
# load labels in tensorflow dataset
train_labels = tf.data.Dataset.list_files('aug_data\\train\\labels\\*.json', shuffle = False)
train_labels = train_labels.map(lambda x: tf.py_function(load_labels,[x],[ tf.uint8 , tf.float16]))
# Since load_labels is a python function ,py_function is used to wrap load_labels function so that it can be applied to elements of tensor dataset, where input is x
# return type is uint8 for class and float16 for bounding box
```

## Deel learning Model

```python
# there will eb one classification model and one regression model
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, Dense, GlobalMaxPooling2D
from tensorflow.keras.applications import VGG16
```

```python
# Model??
```

```python
# download VGG16
vgg = VGG16(include_top=False) # as we don't need last  final layers
```

```python
vgg.summary()
```

```python
# Building model (we are not using sequential here)
def build_model():
    input_layer = Input(shape = (120,120,3))
    
    vgg = VGG16(include_top=False)(input_layer)
    
    #classification Model
    f1 = GlobalMaxPooling2D()(vgg)
    class1 = Dense(2048, activation='relu')(f1)
    class2 = Dense(1, activation= 'sigmoid')(class1)
    
    #regression model
    f2 = GlobalMaxPooling2D()(vgg)
    regress1 = Dense(2048, activation='relu')(f2)
    regress2 = Dense(4, activation= 'sigmoid')(regress1)
    
    facetracker = Model(inputs = input_layer, outputs = [class2 , regress2])
    return facetracker
    
    
```

## Loss and optimizers

```python
# loss function for regression
def localization_loss(y_true, yhat):            
    delta_coord = tf.reduce_sum(tf.square(y_true[:,:2] - yhat[:,:2]))
                  
    h_true = y_true[:,3] - y_true[:,1] 
    w_true = y_true[:,2] - y_true[:,0] 

    h_pred = yhat[:,3] - yhat[:,1] 
    w_pred = yhat[:,2] - yhat[:,0] 
    
    delta_size = tf.reduce_sum(tf.square(w_true - w_pred) + tf.square(h_true-h_pred))
    
    return delta_coord + delta_size

regressloss = localization_loss
```

```python
# Loss function for classification
classloss = tf.keras.losses.BinaryCrossentropy()
```

```python
#testing out loss metrics
X, y = train.as_numpy_iterator().next()
classes, coords = facetracker.predict(X)

print(classloss(y[0], classes))
print(regressloss(y[1], coords))
print(y[0].shape)
print(classes.shape)
```

#### Optimizer

```python
batches_per_epoch = len(train) # one epoch is one full pass through training data
 #how much learning rate drop after each epoch
lr_decay = (1./0.75 -1)/batches_per_epoch # learning rate will cumulatively decay 75% after each epoch,
```

```python
lr_decay
```

```python

opt = tf.keras.optimizers.Adam(learning_rate=0.0001, decay=lr_decay) # newLR = oldLR / ( 1 + deacy*step_nmber)
```

# Train NN

```python
#Subclassing model class from keras
```

```python
class FaceTracker(Model): 
    def __init__(self, facetrackermodel,  **kwargs): # subclass is initialized here(using constructor) by passing initial parameters here the facetracker modle we just created
        super().__init__(**kwargs) # super is used so that this subclass(FaceTracker) properly inherit from parent class(Model from keras) and can be extended here
        # this super().__init__(**kwargs) initilize the parent class
        self.model = facetrackermodel

    def compile(self, opt, classloss, localizationloss, **kwargs):
        super().compile(**kwargs) # calls compile method of parent class, to leverage exixting compilation logic of parent class and can be extended here for handling custom losses
        self.closs = classloss
        self.lloss = localizationloss
        self.opt = opt
    
    def train_step(self, batch, **kwargs): 
        
        # print(f' X : {X.shape}, y: {y.shape}, batch: {batch.shape}')
        X, y = batch
        
        with tf.GradientTape() as tape: # Gradient tape is used to record the forward pass and loss computation
            classes, coords = self.model(X, training=True)
            
            batch_classloss = self.closs(y[0], classes)
            batch_localizationloss = self.lloss(tf.cast(y[1], tf.float32), coords)
            
            total_loss = batch_localizationloss+0.5*batch_classloss
            #tape.gradient is used to compute the gradient of total loss w.r.t model trainable parameters
            grad = tape.gradient(total_loss, self.model.trainable_variables) 
        
        opt.apply_gradients(zip(grad, self.model.trainable_variables)) #weights are updated using the optimizer based on gradients computed 
        
        return {"total_loss":total_loss, "class_loss":batch_classloss, "regress_loss":batch_localizationloss}
    
    def test_step(self, batch, **kwargs): 
        X, y = batch
        
        classes, coords = self.model(X, training=False)
        
        batch_classloss = self.closs(y[0], classes)
        batch_localizationloss = self.lloss(tf.cast(y[1], tf.float32), coords)
        total_loss = batch_localizationloss+0.5*batch_classloss
        
        return {"total_loss":total_loss, "class_loss":batch_classloss, "regress_loss":batch_localizationloss}
        
    def call(self, X, **kwargs): # to use .predict 
        return self.model(X, **kwargs)
```

### Real Time Detection

```python
cap = cv2.VideoCapture(0)
while cap.isOpened():
    _ , frame = cap.read()
    frame = frame[50:500, 50:500,:]
    
    rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    resized = tf.image.resize(rgb, (120,120))
    
    yhat = facetracker.predict(np.expand_dims(resized/255,0))
    sample_coords = yhat[1][0]
    
    if yhat[0] > 0.5: 
        # Controls the main rectangle
        cv2.rectangle(frame, 
                      tuple(np.multiply(sample_coords[:2], [450,450]).astype(int)),
                      tuple(np.multiply(sample_coords[2:], [450,450]).astype(int)), 
                            (255,0,0), 2)
        # Controls the label rectangle
        cv2.rectangle(frame, 
                      tuple(np.add(np.multiply(sample_coords[:2], [450,450]).astype(int), 
                                    [0,-30])),
                      tuple(np.add(np.multiply(sample_coords[:2], [450,450]).astype(int),
                                    [80,0])), 
                            (255,0,0), -1)
        
        # Controls the text rendered
        cv2.putText(frame, 'face', tuple(np.add(np.multiply(sample_coords[:2], [450,450]).astype(int),
                                               [0,-5])),
                    cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,255), 2, cv2.LINE_AA)
    
    cv2.imshow('EyeTrack', frame)
    
    if cv2.waitKey(1) & 0xFF == ord('\r'): # on pressing enter key camera will stop
        break
cap.release()
cv2.destroyAllWindows()
```
